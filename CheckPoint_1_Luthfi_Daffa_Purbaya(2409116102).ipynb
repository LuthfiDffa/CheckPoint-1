```python
# Split genre dan negara
df['genre_split'] = df['listed_in'].str.split(', ')
df['country_split'] = df['country'].str.split(', ')

# Explode genre dan negara untuk membuat baris terpisah untuk setiap genre dan negara
df_genres_countries = df.explode('genre_split').explode('country_split')
```

```python
# Hitung frekuensi genre untuk setiap negara
genre_by_country = df_genres_countries.groupby(['country_split', 'genre_split']).size().unstack(fill_value=0)

# Filter negara dengan konten terbanyak (misalnya, top 10 negara)
top_countries = df_genres_countries['country_split'].value_counts().head(10).index
genre_by_country_top = genre_by_country.loc[top_countries]

# Tampilkan hasil
print(genre_by_country_top)
```
